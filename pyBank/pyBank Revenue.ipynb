```python
import pandas as pd
from pathlib import Path
Path.cwd()
```

```python
file = Path('budget_data.csv')
```

```python
Revenue = pd.read_csv(file)
Revenue
```

```python
# Total months
Revenue_count = Revenue['Date'].count()
Message = f"Total Months: {Revenue_count}"
print(Message)
```

```python
# Net Profit/Loss
Net_Revenue = Revenue['Profit/Losses'].sum()
Message2 = f"Net Profit/Loss: ${Net_Revenue}"
print(Message2)
```

```python
# Average
Average_change = (Revenue['Profit/Losses'] - Revenue['Profit/Losses'].shift(1)).mean()
Average_change
Message3 = f"Average  Change: ${Average_change}"
print(Message3)
```

```python
# The greatest increase in profits (date and amount) over the entire period
Max_change = (Revenue['Profit/Losses'] - Revenue['Profit/Losses'].shift(1)).max()
Max_change
Message4 = f"Greatest Increase in Profits: ${Max_change}"
print(Message4)
```

```python
# The greatest decrease in losses (date and amount) over the entire period
Min_change = (Revenue['Profit/Losses'] - Revenue['Profit/Losses'].shift(1)).min() 
Min_change
Message5 = f"Greatest Decrease in Profits: ${Min_change}"
print(Message5)
```

```python
print(Message)
print(Message2)
print(Message3)
print(Message4)
print(Message5)
```

```python
Revenue['Change'] = Revenue['Profit/Losses'] - Revenue['Profit/Losses'].shift(1)
Revenue.loc[Revenue['Change'] == Max_change]
```
